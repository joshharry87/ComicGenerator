```python
print("Image Stats")


gray_image = gray_imagein



important_Stats = []
important_Stats.append(np.average(gray_image))
important_Stats.append(np.min(gray_image))
important_Stats.append(np.max(gray_image))
important_Stats.append(np.count_nonzero(gray_image))

important_Stats


# important_Stats[0] - important_Stats[1]

print(gray_image.shape[0]/128)
print(gray_image.shape[1]/128)

iteration_step = 128
chunks = []
averages = []
diffs = []
results = []
for x in range(0, int(gray_image.shape[0]/iteration_step)):
    x_run = []

    
    
    for y in range(0, round(gray_image.shape[1]/iteration_step)):
        section = np.array(gray_image[((x)*iteration_step):(iteration_step*(x+1)), ((y)*iteration_step):iteration_step*(y+1)])
        scaled_contrast = (np.square(section / (np.ones(section.shape)*255)))*255
        contrast = abs(scaled_contrast- (np.ones(scaled_contrast.shape)*np.average(scaled_contrast)))
        
        # contrast = abs(section- (np.ones(section.shape)*np.average(section)))
        inverted =( section - (np.ones(section.shape)*255))*-1
        # averages[f'{x}_{y}'] = np.average(gray_image[((x)*iteration_step):(iteration_step*x), ((y)*iteration_step):iteration_step*y])
        chunks.append((inverted, f'{x}_{y}', section))
        averages.append(np.sum(section))
        diffs.append(np.sum(contrast))
        x_run.append(np.sum(contrast))
        
        
    results.append(x_run)
    

```

```python
np.array(results).shape
```

```python
window = 6

results = np.array(results)

id_groupings = []
centre = []

for x in range(0,results.shape[0]):
    for y in range(0,results.shape[1]):
        section = np.sum(results[max(x-(int(window/2)), 
                                     0):min(x+(int(window/2)), results.shape[0]), 
                                 max(0,y-(int(window/2))):min(results.shape[1],y+(int(window/2)))])
        
        id_groupings.append(section)
        
        centre.append(f'{x}_{y}')
        
        
        
    
```

```python
id_groupings.index(np.array(id_groupings).max())

thresh2 = 0.65

interesting_groups = []

for x in id_groupings:
    if x > (np.array(id_groupings).max()*thresh2):
        interesting_groups.append((x, id_groupings.index(x)))
```
