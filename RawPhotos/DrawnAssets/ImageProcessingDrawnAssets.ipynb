```python
print("Image Stats")


# gray_image = gray_imagein

gray_image =quickTest

important_Stats = []
important_Stats.append(np.average(gray_image))
important_Stats.append(np.min(gray_image))
important_Stats.append(np.max(gray_image))
important_Stats.append(np.count_nonzero(gray_image))

important_Stats


# important_Stats[0] - important_Stats[1]

print(gray_image.shape[0]/128)
print(gray_image.shape[1]/128)

iteration_step = 64
chunks = []
averages = []
diffs = []
results = []
for x in range(0, int(gray_image.shape[0]/iteration_step)):
    x_run = []

    
    
    for y  in range(0, round(gray_image.shape[1]/iteration_step)):
        section = np.array(gray_image[((x)*iteration_step):(iteration_step*(x+1)), ((y)*iteration_step):iteration_step*(y+1)])
        scaled_contrast = (np.square(section / (np.ones(section.shape)*255)))*255
        contrast = abs(scaled_contrast- (np.ones(scaled_contrast.shape)*np.average(scaled_contrast)))
        
        # contrast = abs(section- (np.ones(section.shape)*np.average(section)))
        inverted =( section - (np.ones(section.shape)*255))*-1
        # averages[f'{x}_{y}'] = np.average(gray_image[((x)*iteration_step):(iteration_step*x), ((y)*iteration_step):iteration_step*y])
        chunks.append((inverted, f'{x}_{y}', section))
        averages.append(np.sum(section))
        diffs.append(np.sum(contrast))
        x_run.append(np.sum(contrast))
        
        
    results.append(x_run)
    

```

```python
window = 6

results = np.array(results)

id_groupings = []
centre = []

for x in range(0,results.shape[0]):
    for y in range(0,results.shape[1]):
        section = np.sum(results[max(x-(int(window/2)), 
                                     0):min(x+(int(window/2)), results.shape[0]), 
                                 max(0,y-(int(window/2))):min(results.shape[1],y+(int(window/2)))])
        
        id_groupings.append(section)
        
        centre.append(f'{x}_{y}')
        
        
        
    
```

```python
id_groupings.index(np.array(id_groupings).max())

thresh2 = 0.95

interesting_groups = []

for x in id_groupings:
    if x > (np.array(id_groupings).max()*thresh2):
        interesting_groups.append((x, id_groupings.index(x)))
```

```python
def group_interesting_tiles(tiles, chunksin, iteration_step):
    groups = []
    
    
    group = {
        'start_x' : 0,
        'start_y' : 0,
        'end_x' : 0,
        'end_y' : 0,
        'tile' : False
    }

    
    for tile in tiles:
        new_group_flag = False
        tile_flag = True
        print(tile)
        x , y = int(chunksin[tile][1].split("_")[0])*iteration_step, int(chunksin[tile][1].split("_")[1])*iteration_step
        for collection in groups:
            if (collection['end_y'] == y and (x >= collection['start_x'] and x <= collection['end_x'])  ):
                collection['end_y'] = y +iteration_step
                tile_flag = False         
            elif (x == collection['end_x'] and y + iteration_step == collection['end_y']):
                collection['end_x'] = x + iteration_step
                tile_flag = False
            # elif  (collection['end_y'] >= y +iteration_step and collection['start_y'] <= y  
            #        and (x >= collection['start_x'] and x <= collection['end_x'])  ):
            #     print("exists")
            #     tile_flag = False
            elif  (collection['end_y'] >= y  and collection['start_y'] <= y  
                   and (x >= collection['start_x'] and x <= collection['end_x'])  ):
                print("exists")
                tile_flag = False
            
            else:
                new_group_flag = True
        
        
        if (new_group_flag and tile_flag):
                
            new_group =  {
                            'start_x' : x-iteration_step,
                            'start_y' : y-iteration_step,
                            'end_x' : x+iteration_step,
                            'end_y' :  y +iteration_step,
                        } 
            groups.append(new_group)

        if len(groups) == 0:
            new_group =  {
                                'start_x' : x-iteration_step,
                                'start_y' : y-iteration_step,
                                'end_x' : x+iteration_step,
                                'end_y' :  y +iteration_step,
                            } 
            print(group)
            groups.append(new_group)
        # print(groups)
        tile_flag == True
    for collection in groups:
        collection['end_y'] = collection['end_y'] + iteration_step
        collection['end_x'] = collection['end_x'] + iteration_step
    remove = []
    for i in range(0, len(groups)):
        for ii in range(0, len(groups)):
            if groups[i]['end_y'] >= groups[ii]['start_y'] and groups[i]['start_y'] >= groups[ii]['start_y'] and groups[i]['start_y'] <= groups[ii]['end_y'] and i != ii:
                groups[i]['end_y'] =  max(groups[i]['end_y'], groups[ii]['end_y'])
                groups[i]['start_y'] =  min(groups[i]['start_y'], groups[ii]['start_y'])
                groups[i]['end_x'] =  max(groups[i]['end_x'], groups[ii]['end_x'])
                groups[i]['start_x'] =  min(groups[i]['start_x'], groups[ii]['start_x'])
                groups[ii]['end_y'] =  max(groups[i]['end_y'], groups[ii]['end_y'])
                groups[ii]['start_y'] =  min(groups[i]['start_y'], groups[ii]['start_y'])
                groups[ii]['end_x'] =  max(groups[i]['end_x'], groups[ii]['end_x'])
                groups[ii]['start_x'] =  min(groups[i]['start_x'], groups[ii]['start_x'])
                print("overlap y")
                
            

     
    unique = []
    
    for item in groups:
        try:
            if iteration_step < 100:
                item["end_x"] = item["end_x"] + iteration_step
            unique.index(item)
        except:

            unique.append(item)
    
    
    
    
    return  unique
                
            
        # cv2.imwrite(f'b_tile_pixel_{tile}_x{x}_y{y}.jpg', chunksin[tile][2])
        
        
```
